```python
# ─── CELL 4 — Path Setup + Data Routing ────────────────────────────
import os, sys, shutil

REPO_DIR = '/kaggle/working/STELLARIS-DNet'
os.chdir(REPO_DIR)
sys.path.insert(0, REPO_DIR)

# ── HTRU2 ──────────────────────────────────────────────────────────
# Kaggle dataset: https://www.kaggle.com/datasets/colearninglounge/predicting-pulsar-star

HTRU2_KAGGLE_CANDIDATES = [
    '/kaggle/input/htru2-pulsar-dataset/HTRU_2.csv',
    '/kaggle/input/htru2/HTRU_2.csv',
    '/kaggle/input/predicting-pulsar-star/pulsar_stars.csv',
]

def find_htru2_file():
    # 1️⃣ Fast path (your original assumptions)
    for p in HTRU2_KAGGLE_CANDIDATES:
        if os.path.exists(p):
            return p

    # 2️⃣ Robust fallback (auto-scan Kaggle input)
    for root, dirs, files in os.walk('/kaggle/input'):
        for f in files:
            if f in ['HTRU_2.csv', 'pulsar_stars.csv']:
                return os.path.join(root, f)

    return None


htru2_src = find_htru2_file()

if htru2_src:
    os.makedirs('data/module1', exist_ok=True)
    dst = 'data/module1/HTRU_2.csv'

    if not os.path.exists(dst):
        shutil.copy(htru2_src, dst)

    print(f'✅ HTRU2 found at: {htru2_src}')
    print(f'📦 Copied → {dst}')
else:
    print('❌ HTRU2 not found — dataset missing or incorrectly attached')
    print('   👉 Attach dataset from:')
    print('   https://www.kaggle.com/datasets/colearninglounge/predicting-pulsar-star')

# ── Checkpoint + Log dirs ───────────────────────────────────────────
for d in ['checkpoints/module1', 'logs/module1', 'data/module1']:
    os.makedirs(d, exist_ok=True)
    print(f'📁 {d}')

print(f'\nWorking dir : {os.getcwd()}')
print(f'HTRU2 path  : {os.path.abspath("data/module1/HTRU_2.csv")}')
```

```python
# ─── CELL 12 — Display Training Curves ─────────────────────────────
import os, json
import matplotlib.pyplot as plt
from module1.config import LOG_DIR, EXPERIMENT_TAG

def load_history(name):
    path = os.path.join(LOG_DIR, f'{name}_history.json')
    if not os.path.exists(path):
        return None
    with open(path) as f:
        return json.load(f)

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('Module 1 — Training Curves', fontsize=14)

names    = ['mlp', 'cnn', 'ae']
titles   = ['MLP (HTRU2)', 'CNN (Subtypes)', 'Autoencoder']
colors   = [('steelblue', 'darkorange'), ('seagreen', 'tomato'),
            ('purple', 'crimson')]

for ax, name, title, (c1, c2) in zip(axes, names, titles, colors):
    tag  = EXPERIMENT_TAG
    hist = load_history(f'{name}_{tag}') or load_history(name)
    if hist is None:
        ax.text(0.5, 0.5, f'No history for {name}',
                ha='center', va='center')
        ax.set_title(title); continue
    epochs = range(1, len(hist['train_loss']) + 1)
    ax.plot(epochs, hist['train_loss'], color=c1, lw=1.5, label='Train')
    ax.plot(epochs, hist['val_loss'],   color=c2, lw=1.5, label='Val')
    ax.set_title(title); ax.set_xlabel('Epoch'); ax.set_ylabel('Loss')
    ax.legend(); ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# ─── CELL 13 — Baseline vs Enhanced Comparison ─────────────────────
import os, json
import matplotlib.pyplot as plt
from module1.config import LOG_DIR, EXPERIMENT_TAG

b_path = os.path.join(LOG_DIR, 'mlp_baseline_history.json')
e_path = os.path.join(LOG_DIR, f'mlp_{EXPERIMENT_TAG}_history.json')

if os.path.exists(b_path) and os.path.exists(e_path):
    with open(b_path) as f: baseline = json.load(f)
    with open(e_path) as f: enhanced = json.load(f)

    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    fig.suptitle('MLP — Baseline vs Enhanced', fontsize=13)

    for ax, metric, ylabel in [
        (axes[0], 'val_loss', 'Val Loss'),
        (axes[1], 'val_f1',   'Val F1'),
    ]:
        ep_b = range(1, len(baseline.get(metric, [])) + 1)
        ep_e = range(1, len(enhanced.get(metric, [])) + 1)
        if baseline.get(metric):
            ax.plot(ep_b, baseline[metric], '--', lw=1.5,
                    color='steelblue', label='Baseline (BCE)')
        if enhanced.get(metric):
            ax.plot(ep_e, enhanced[metric], lw=2,
                    color='darkorange', label=f'Enhanced (Focal+Physics)')
        ax.set_xlabel('Epoch'); ax.set_ylabel(ylabel)
        ax.set_title(ylabel); ax.legend(); ax.grid(alpha=0.3)

    plt.tight_layout()
    plt.show()

    # Numeric comparison table
    def last(d, k): return d[k][-1] if d.get(k) else float('nan')
    print('\n' + '=' * 52)
    print(f'{"Metric":<18} {"Baseline":>10} {"Enhanced":>10} {"Δ":>10}')
    print('=' * 52)
    for metric, label in [
        ('val_loss',      'Val Loss'),
        ('val_acc',       'Val Accuracy'),
        ('val_f1',        'Val F1'),
        ('val_precision', 'Val Precision'),
        ('val_recall',    'Val Recall'),
    ]:
        b = last(baseline, metric); e = last(enhanced, metric)
        d = e - b
        arrow = '↑' if d > 0 else ('↓' if d < 0 else '=')
        print(f'{label:<18} {b:>10.4f} {e:>10.4f} {d:>+9.4f}{arrow}')
    print('=' * 52)
else:
    print('⚠️  Comparison data not found.')
    print(f'   Expected: {b_path}')
    print(f'   Expected: {e_path}')
```

```python
# --- CELL 4 - Code Snippet Extraction ------------------------------------
# Extracts exactly 3-5 lines per function/class for PPT slides
# PDF alignment:
#   extract_fft_features -> ASR Feature Extraction analogy
#   augment_profile      -> Image Classification Data Augmentation
#   MultiHeadTemporalAttention -> DL-ASR Transformer component
#   conv_encoder block   -> DL-ASR CNN local pattern extraction

import os

SNIPPETS = {

    # Topic 2: ASR Feature Extraction analogy
    'topic2_asr_fft_feature_extraction.py': {
        'pdf_topic': 'ASR Feature Extraction - MFCC/Spectrogram analogy (PDF Page 8)',
        'slide_note': 'Show alongside ASR MFCC diagram from PDF slide 8',
        'code': (
            "def extract_fft_features(profile, n_bins=FFT_BINS):\n"
            "    # Pulse profile -> frequency domain (like MFCC in ASR)\n"
            "    fft_mag = np.abs(np.fft.rfft(profile))[:n_bins]\n"
            "    fft_log = np.log1p(fft_mag)                      # log scale\n"
            "    return (fft_log / fft_log.max()).astype(np.float32)  # normalize\n"
        ),
    },

    # Topic 1: Image Classification -> Data Augmentation
    'topic1_image_classification_augmentation.py': {
        'pdf_topic': 'Image Classification - Data Augmentation technique (PDF Page 4)',
        'slide_note': 'Show alongside Data Augmentation row from PDF page 4 table',
        'code': (
            "def augment_profile(profile):\n"
            "    # Augments pulse profile: shift + scale + noise\n"
            "    # Mirrors image augmentation (flip, rotate, distort) from PDF\n"
            "    profile = np.roll(profile, np.random.randint(-SHIFT_MAX, SHIFT_MAX))\n"
            "    profile = profile * np.random.uniform(SCALE_MIN, SCALE_MAX)\n"
            "    return inject_noise(profile, NOISE_STD)\n"
        ),
    },

    # Topic 3: DL-ASR -> Transformer component
    'topic3_dl_asr_transformer_attention.py': {
        'pdf_topic': 'DL-Based ASR - Transformer: long-range dependencies (PDF Page 10)',
        'slide_note': 'Show alongside Transformer row from PDF page 10 table',
        'code': (
            "class MultiHeadTemporalAttention(nn.Module):\n"
            "    # Self-attention over pulse time axis\n"
            "    # Mirrors Transformer role in DL-ASR (PDF page 10)\n"
            "    def __init__(self, embed_dim, num_heads, dropout=0.1):\n"
            "        super().__init__()\n"
            "        self.attn = nn.MultiheadAttention(\n"
            "            embed_dim, num_heads, dropout=dropout, batch_first=True)\n"
            "        self.norm = nn.LayerNorm(embed_dim)\n"
        ),
    },

    # Topic 3: DL-ASR -> CNN local pattern extraction
    'topic3_dl_asr_cnn_local_patterns.py': {
        'pdf_topic': 'DL-Based ASR - CNNs: extract local patterns in spectrograms (PDF Page 10)',
        'slide_note': 'Show alongside CNN row from PDF page 10 table',
        'code': (
            "# PulsarCNN - conv encoder (local pulse pattern extraction)\n"
            "# Mirrors CNN role in ASR: extracts local patterns from signal\n"
            "# PDF Page 10: CNNs -> Extract local patterns in spectrograms\n"
            "Conv1d(1,   64,  kernel_size=7, padding=3)  # broad features\n"
            "Conv1d(64,  128, kernel_size=5, padding=2)  # mid-scale features\n"
            "Conv1d(128, 256, kernel_size=3, padding=1)  # fine-grained features\n"
        ),
    },

    # Topic 1: Image Classification -> CNN full pipeline
    'topic1_image_classification_cnn_pipeline.py': {
        'pdf_topic': 'Image Classification - CNN: auto feature extraction + classify (PDF Page 4)',
        'slide_note': 'Maps to CNN row in PDF page 4 techniques table',
        'code': (
            "# PulsarCNN - mirrors Image Classification pipeline (PDF page 3)\n"
            "# Step 1: Input -> Step 2: Feature Extraction -> Step 3: Classification\n"
            "x = self.conv_encoder(x_time)        # Step 2: Feature Extraction\n"
            "x = self.temporal_attn(x)            # Long-range dependencies\n"
            "x = self.pool(x).squeeze(-1)         # Global pooling -> fixed vector\n"
            "return self.head(self.time_proj(x))  # Step 3: Classification output\n"
        ),
    },
}

# --- Save all snippets ---
print('── Code Snippet Extraction ──')
for filename, content in SNIPPETS.items():
    path = os.path.join(DIRS['snippets'], filename)
    with open(path, 'w') as f:
        f.write('# PDF Topic  : ' + content['pdf_topic'] + '\n')
        f.write('# Slide Note : ' + content['slide_note'] + '\n')
        f.write('# ' + '-' * 60 + '\n')
        f.write(content['code'])
    lines = content['code'].count('\n')
    print(f'  OK  {filename}  ({lines} lines)')

print()
print(str(len(SNIPPETS)) + ' snippets saved -> ' + DIRS['snippets'])
```
